```python
import pandas as pd

# Seleccionamos las variables relevantes de ambos conjuntos
# (Usamos merge para asegurar que tengan los mismos nombres de columnas si los quieres unir)
vars_leche = pd.DataFrame()

vars_leche['PRECIO_LECHE'] = leche_dolar['precio promedio por litro']
vars_leche['IPC'] = leche_dolar['ipc']
vars_leche['D√ìLAR'] = leche_dolar['PRECIO_DOLAR']
vars_leche['GASOLINA_CORRIENTE'] = leche_dolar_gas['PRECIO_PROMEDIO_CORRIENTE']
vars_leche['GASOLINA_EXTRA'] = leche_dolar_gas['PRECIO_PROMEDIO_EXTRA']
vars_leche['BIODI√âSEL'] = leche_dolar_gas['PRECIO_PROMEDIO_BIODIESEL']

# Calcular correlaci√≥n de Pearson
correlaciones = vars_leche.corr(method='pearson')

# Extraer solo la fila correspondiente al precio de la leche
correlacion_leche = correlaciones.loc['PRECIO_LECHE'].drop('PRECIO_LECHE')

# Mostrar resultados ordenados
correlacion_leche = correlacion_leche.sort_values(ascending=False)

# Mostrar resultados
print("üìä Correlaci√≥n (coeficiente de Pearson) respecto al precio de la leche:\n")
display(correlacion_leche.to_frame('Coeficiente de correlaci√≥n').style.background_gradient(cmap='RdYlGn', vmin=-1, vmax=1))
```

```python
import pandas as pd

# ===============================
# 1Ô∏è‚É£ Selecci√≥n de variables base
# ===============================

vars_leche = pd.DataFrame()
vars_leche['A√ëO'] = leche_dolar['a√±o']
vars_leche['MES'] = leche_dolar['mes']
vars_leche['PRECIO_LECHE'] = leche_dolar['precio promedio por litro']
vars_leche['IPC'] = leche_dolar['ipc']
vars_leche['D√ìLAR'] = leche_dolar['PRECIO_DOLAR']

# Agregar combustibles (solo a√±os donde existen)
vars_leche['GASOLINA_CORRIENTE'] = leche_dolar_gas['PRECIO_PROMEDIO_CORRIENTE']
vars_leche['BIODI√âSEL'] = leche_dolar_gas['PRECIO_PROMEDIO_BIODIESEL']

# =================================
# 2Ô∏è‚É£ Dividir los periodos temporalmente
# =================================

# Definir punto de quiebre del IPC (febrero de 2019)
punto_quiebre = (2019, 2)

pre_pandemia = vars_leche[(vars_leche['A√ëO'] < 2019) | ((vars_leche['A√ëO'] == 2019) & (vars_leche['MES'] <= 2))]
post_pandemia = vars_leche[(vars_leche['A√ëO'] > 2019) | ((vars_leche['A√ëO'] == 2019) & (vars_leche['MES'] > 2))]

# ============================
# 3Ô∏è‚É£ Funci√≥n para calcular correlaciones
# ============================

def correlacion_periodo(df, variables, titulo):
    subset = df[variables].dropna()
    corr = subset.corr(method='pearson')
    correlacion_leche = corr.loc['PRECIO_LECHE'].drop('PRECIO_LECHE').sort_values(ascending=False)
    
    print(f"\nüìä {titulo}")
    display(
        correlacion_leche
        .to_frame('Coeficiente de correlaci√≥n')
        .style.background_gradient(cmap='RdYlGn', vmin=-1, vmax=1)
    )

# =================================
# 4Ô∏è‚É£ Calcular correlaciones por escenario
# =================================

# Escenario 1: Leche vs D√≥lar (todos los a√±os)
correlacion_periodo(vars_leche, ['PRECIO_LECHE', 'D√ìLAR'], "Leche vs D√≥lar (2013‚Äì2024)")

# Escenario 2: Leche vs D√≥lar + IPC (pre-pandemia)
correlacion_periodo(pre_pandemia, ['PRECIO_LECHE', 'D√ìLAR', 'IPC'], "Leche vs D√≥lar + IPC (Pre-pandemia, hasta feb 2019)")

# Escenario 3: Leche vs D√≥lar + IPC (post-pandemia)
correlacion_periodo(post_pandemia, ['PRECIO_LECHE', 'D√ìLAR', 'IPC'], "Leche vs D√≥lar + IPC (Post-pandemia, desde mar 2019)")

# Escenario 4: Leche vs D√≥lar + IPC + combustibles (sin gasolina extra, solo a√±os con datos)
correlacion_periodo(
    vars_leche.dropna(subset=['GASOLINA_CORRIENTE', 'BIODI√âSEL']),
    ['PRECIO_LECHE', 'D√ìLAR', 'IPC', 'GASOLINA_CORRIENTE', 'BIODI√âSEL'],
    "Leche vs D√≥lar + IPC + Gasolinas (2020‚Äì2024)"
)
```
