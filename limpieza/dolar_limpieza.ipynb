```python
import pandas as pd
```

```python
# Cargar el archivo CSV
dolar = pd.read_csv("../Dolar/wilkinsonpc-dolar-historico-2010-01-03-al-2025-12-03.csv")  # Ajusta la ruta si está en otra carpeta

# Convertir la columna de fecha a tipo datetime
dolar['Fecha'] = pd.to_datetime(dolar['Fecha'])

# Crear columnas de año y mes
dolar['AÑO'] = dolar['Fecha'].dt.year
dolar['MES'] = dolar['Fecha'].dt.month
```

```python
# Calcular el promedio mensual del precio del dólar
dolar_promedios = (
    dolar.groupby(['AÑO', 'MES'], as_index=False)['Precio']
         .mean()
         .rename(columns={'Precio': 'PRECIO_PROMEDIO'})
)

# Mostrar los primeros resultados
dolar_promedios.head()
```

```python
# Guardar en un nuevo CSV
dolar_promedios.to_csv("../Dolar/promedio_mensual_dolar.csv", index=False)

print("✅ Archivo guardado en ../Dolar/promedio_mensual_dolar.csv")
```
