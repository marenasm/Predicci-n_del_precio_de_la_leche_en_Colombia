```python
import pandas as pd
```

```python
# Archivo base (precios por municipio)
base = pd.read_csv("../Data/leche_yarumal_ipc_limpio.csv")

# Promedios mensuales del dólar
dolar = pd.read_csv("../Dolar/promedio_mensual_dolar.csv")

# Promedios mensuales de gasolina por producto
gasolina = pd.read_csv("../Gasolina/promedios_mensuales_productos.csv")
```

```python
print("Base municipios:", base.columns.tolist())
print("Dólar:", dolar.columns.tolist())
print("Gasolina:", gasolina.columns.tolist())
```

```python
Base_municipios: ['AÑO', 'MES', 'nombre departamento', 'código departamento', 'nombre municipio', 'código municipio', 'precio promedio por litro', 'ipc']
Dólar: ['AÑO', 'MES', 'PRECIO_PROMEDIO']
Gasolina: ['AÑO', 'MES_NUM', 'PRECIO_PROMEDIO_BIODIESEL', 'PRECIO_PROMEDIO_CORRIENTE', 'PRECIO_PROMEDIO_EXTRA']
```

```python
# Asegurar compatibilidad de nombres de columnas
dolar.rename(columns={'MES': 'MES_NUM', 'PRECIO_PROMEDIO': 'PRECIO_DOLAR'}, inplace=True)

# Unir por año y mes
base_dolar = pd.merge(base, dolar, left_on=['año', 'mes'], right_on=['AÑO', 'MES_NUM'], how='left')

# Eliminar columnas duplicadas de merge
base_dolar.drop(columns=['AÑO', 'MES_NUM'], inplace=True)

# Mostrar resultado
base_dolar.head()

```

```python
base_dolar.tail()
```

```python
base_dolar.to_csv("../Data/base_con_dolar.csv", index=False)
print("✅ Archivo guardado: ../Data/base_con_dolar.csv")
```

```python
# Asegurar consistencia de nombres
gasolina.rename(columns={'MES_NUM': 'mes', 'AÑO': 'año'}, inplace=True)

# Unir por año y mes, pero solo donde haya datos de gasolina
base_dolar_gas = pd.merge(base_dolar, gasolina, on=['año', 'mes'], how='inner')

# Mostrar las primeras filas
base_dolar_gas.head()
```

```python
base_dolar_gas.to_csv("../Data/base_con_dolar_gasolina.csv", index=False)
print("✅ Archivo guardado: ../Data/base_con_dolar_gasolina.csv")
```
